```python
import pandas as pd

customers = pd.read_csv("customers.csv")
products = pd.read_csv("products.csv")
orders = pd.read_csv("orders.csv")
payments = pd.read_csv("payments.csv")
```

```python
customerorder = pd.merge(customers, orders, left_on= 'customer_id', right_on= 'customer_id', how= 'outer')
customerorder
```

```python
customerproducts = pd.merge(customerorder, products, left_on= 'product_id', right_on= 'product_id', how= 'outer')
customerproducts
```

```python
orderdetails = pd.merge(customerproducts, payments, left_on= 'order_id', right_on= 'order_id', how= 'outer')
orderdetails
```

```python
# orderdetails.drop(columns='category')
```

```python
orderdetails.insert(1, 'Full Name', 'N/A') 
orderdetails['Full Name'] = orderdetails['first_name'] + ' ' + orderdetails['last_name']
orderdetails.drop(columns = ['first_name', 'last_name'], inplace=True)
orderdetails.drop(columns= 'category', inplace=True)
orderdetails.drop(columns= ['payment_id', 'product_id', 'order_id', 'customer_id'], inplace=True)
orderdetails
```

```python
orderdetails['Amount'] = orderdetails['quantity'] * orderdetails['price']
orderdetails
```

```python
orderdetails.groupby(['payment_status']).agg({'quantity': 'sum', 'price': 'sum'})
```

```python
# orderdetails.loc[len(orderdetails)] = ["Louis", 'New York', 'USA', 10, '2026-03-03', 'Laptop', 1200, 'Cash', 'Completed', 12000]
orderdetails.to_csv("orderdetails.csv", index=False)
```
