```python
# coffee.insert(3, 'Selling Price', 5.99)
# coffee.insert(4, 'Revenue', 1)
# coffee.insert(3, 'Production Cost', 0.99)
# coffee.insert(6, 'Profit', 0)

coffee['Selling Price'] = np.where(coffee['Coffee Type']  == 'Espresso', 3.99, 5.99)
coffee["Revenue"] = coffee["Units Sold"] * coffee["Selling Price"]
coffee["Production Cost"] = np.where(coffee['Coffee Type'] == 'Espresso', 0.99, 1.99)
coffee["Profit"] = coffee["Revenue"] - (coffee["Units Sold"] * coffee["Production Cost"])
coffee
```

```python
bios['born_datetime'] = pd.to_datetime(bios['born_date'])
bios['born_year'] = bios['born_datetime'].dt.year
# bios[['name', 'born_year']]
bios.head()
```

```python
bios['weight'] = bios['weight_kg'].apply(lambda x: 'Light' if x < 60 else ('Heavy' if x > 75 else 'Normal'))
bios.head(25)
```

```python
ahmed = bios[bios['name'].str.contains("Ahmed", case=False)]
saad = bios[bios['name'].str.contains("Saad", case=False)]
ahmed_saad = pd.concat([ahmed, saad])
ahmed_saad
```
